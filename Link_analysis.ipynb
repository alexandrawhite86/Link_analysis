```python
## import libraries
import pandas as pd
```

```python
## import link nyc location data
link=pd.read_csv("raw_data/LinkNYC_Locations.csv")
link.head(10)
```

```python
link.info()
```

```python
##drop the columns you don't need 
link_new=link.drop(columns=["Council District","Project Status","Smallest PPT","Link Smoke Tested and Activated (A)","Neighborhood Tabulation Area (NTA)","Building Identification Number (BIN)","Borough-Block-Lot (BBL)", "Census Tract (CT)", "Cross Street 1", "Cross Street 2", "IxN Corner","GF Permit Application ID", "Location", "X Coordinate","Y Coordinate"])
```

```python
##new sheet with dropped columns 
link_new.info()
```

```python
## How many links does each community board have
link_new["Community Board"].value_counts()
```

```python
#create new dataframe with count values 
link_new_count = link_new.value_counts("Community Board").rename_axis("Community Board").reset_index(name='#_of_links')
print (link_new_count)
```

```python
link_new_count.info()
```

```python
link_new_count.head(40)
```

```python
##rename community board to cd_number 
link_new_count = link_new_count.rename(columns={'Community Board': 'cd_number'})
```

```python
##Final look of dataframe
link_new_count.head(40)
```

```python
##import internet access csv
internet=pd.read_csv("raw_data/Internet_Access.csv")
```

```python
internet.head(20)
```

```python
internet.info()
```

```python
##rename FIPS to cd_number, location to cd_name
internet = internet.rename(columns={'Fips': 'cd_number', 'Location': 'cd_name'})
```

```python
internet.info()
```

```python
##filter out what you need
filter1 = internet ["DataFormat"] == "Percent"
filter2 = internet ["TimeFrame"] == 2019
##filter

internet_new = internet[filter1 & filter2]

```

```python
internet_new.head(10)
```

```python
internet_new['cd_number'].unique()
```

```python
link_new_count['cd_number'].unique()
```

```python
internet_new.info()
```

```python
link_new_count.info()
```

```python
## join internet_new, area_new and  and link_new_count dataframes together into a dataframe called link_nyc
#join_list=[internet_new, link_new_count] ## 2 dataframes to join - this didnt work  
#link_nyc = pd.concat(objs=[internet_new, link_new_count], sort =True) - this didnt work 
link_nyc = internet_new.merge(link_new_count, left_on= 'cd_number', right_on= 'cd_number')
```

```python
link_nyc.info()
```

```python
link_nyc.head(10)
```

```python
missing_cd= set(internet_new['cd_number'].unique())- set(link_new_count['cd_number'].unique())
```

```python
filter3 = internet_new ["cd_number"].isin(missing_cd)
##filter
missing_cd_data= internet_new[filter3]
```

```python
missing_cd_data
```

```python
join_list=[link_nyc, missing_cd_data] ## 2 dataframes to join 
link_nyc_new = pd.concat(objs= join_list, sort =True)
```

```python
link_nyc_new
```

```python
link_nyc_no_na= link_nyc_new.fillna(0)
```

```python
link_nyc_no_na
```

```python
link_nyc_no_na.sort_values(by="Data")
```

```python
##rename data to % of households w/o internet access, time frame to year
link_nyc_no_na = link_nyc_no_na.rename(columns={'Data': '%_of_households_w/o_internet', 'TimeFrame': 'year'})
##drop data format column
link_nyc_no_na = link_nyc_no_na.drop(columns=["DataFormat"])
```

```python
link_nyc_no_na
```

```python
## Get sum of # of links in NYC
link_nyc_no_na["#_of_links"].sum()
#these are the links installed not the links that are active only 1820 active links but 1869 installed, wondering why they are 49 links installed that are not active 
```

```python
##sort values by # of links 
link_nyc_no_na.sort_values(by="#_of_links")
## Areas w/ the lowest percentage of people without internet access tend to have the most amount of links 
## Murray hill has the most amount of links but only 4% of households have no internet access in comparison to Bourough Park which has no no link terminals and 33% of households don't have internet access 
```

```python
##sort values by households w/o internet 
link_nyc_no_na.sort_values(by="%_of_households_w/o_internet")
##there are still some areas with low amounts of households w/o internet access that don't have links: Howard Beach, Canarsie
## But Murray hill has the lowest amount of households without internet access and it has the most number of links 
```
